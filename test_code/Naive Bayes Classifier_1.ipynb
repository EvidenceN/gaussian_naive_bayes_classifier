```
# calculate mean, standard deviation, exponent, and get pi

# first part of equation. 

# mean and standard deviation calculation
# https://numpy.org/doc/stable/reference/routines.statistics.html#averages-and-variances

# pi and base e exponent are constants
# https://numpy.org/doc/stable/reference/constants.html#numpy.e

# square root function
# https://numpy.org/doc/stable/reference/generated/numpy.sqrt.html

# alternatives are 
# https://docs.python.org/3/library/math.html
# https://docs.python.org/3/library/statistics.html#statistics.stdev

# x(i) = each observation = each row
```

```
import numpy as np
```

```
# what is being calculated is the probability of each observation given
# the y values. So, the result of the calculation should be a dataset

# the fit method is calculating the probability of each observation given y
# the predict method is using the x values to predict y

# how it is done
# https://en.wikipedia.org/wiki/Naive_Bayes_classifier#Examples
def gnb(x, y):
  # y standard deviation
  y_dev = np.std(y)
  # standard deviation raised to the power 2 = variance
  y_var = np.var(y)
  # y mean value
  y_mean = np.mean(y)
  # natural log
  e = np.e
  # pi
  pi = np.pi
  # first part of the equation
  # 1 divided by the sqrt of 2 * pi * y_variance
  equation_1 = 1/(np.sqrt(2 * pi * y_var))
  
  # second part of equation implementation
  # denominator of equation
  denom = 2 * y_var

  # numerator calculation

  # make sure our x_dataframe is a numpy array to ensure 
  # ease of subtraction and calculations
  x = np.array(x)

  # guassian naive bayes equation is applied to each observation
  # create a new x dataframe that has just the numerator
  new_x = []
  for i in range(len(x)):
    # numerator
    numerator = (x[i] - y_mean) ** 2 
    # the exponent
    expo = np.exp(-(numerator/denom))
    p = equation_1 * expo
    new_x.append(p)

  return new_x

```
